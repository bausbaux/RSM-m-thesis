```python
import re

# Function to extract company name from firm_link
def extract_company_name(link):
    # Remove full URL prefix if present
    link = link.split('Reviews/')[1] if 'Reviews/' in link else link
    # Extract text before "-Reviews"
    match = re.match(r'([^-]+(?:-[^-]+)*)-Reviews', link)
    return match.group(1).replace('-', ' ') if match else None

# Apply function to create new column
df['company_name'] = df['firm_link'].apply(extract_company_name)

# Count the number of reviews per company
company_counts = df['company_name'].value_counts().reset_index()
company_counts.columns = ['company_name', 'review_count']

# Optional: Display top 10 companies
print(company_counts.head(10))
```

```python
#removing the missing company rows as this is needed to infer whether or not a company is in the EU or not as well as company size for control variable
df = df.dropna(subset=['company_name'])
```

```python
sentiment_mapping = {
    'v': 1,    # Positive
    'r': 0.5,  # Mild
    'x': -1,   # Negative
    'o': 0     # No opinion
}

# Map the columns (CEO Approval and Business Outlook and recommend) to numeric sentiment values
df['CEO_Approval_Sentiment'] = df['CEO Approval'].map(sentiment_mapping)
df['Business_Outlook_Sentiment'] = df['Business Outlook'].map(sentiment_mapping)

# Check the first few rows to see the transformed data
print(df[['CEO Approval', 'CEO_Approval_Sentiment', 'Business Outlook', 'Business_Outlook_Sentiment']].head())
```

```python
df['Recommmend_sentiment'] = df['Recommend'].map(sentiment_mapping)
```

```python
# Extract unique company names
company_names = df['company_name'].dropna().unique()

# Convert to a DataFrame
company_names_df = pd.DataFrame(company_names, columns=['company_name'])

# Save to CSV
company_names_df.to_csv('unique_company_names.csv', index=False)
```

```python
# Drop rows with missing ratings just in case
df_clean = df.dropna(subset=['rating'])

# Count ratings per company
rating_counts = df_clean.groupby('company_name')['rating'].count()

# Filter companies with > 1000 ratings
over_1000_ratings = rating_counts[rating_counts > 1000]

# Convert to DataFrame for display or export
over_1000_ratings_df = over_1000_ratings.reset_index()
over_1000_ratings_df.columns = ['company_name', 'rating_count']

# View the result
#print(over_1000_ratings_df)
```

```python
total_ratings = over_1000_ratings_df['rating_count'].sum()
print(f"Total number of ratings from companies with over 1000 ratings: {total_ratings}")
```

```python
# Load enriched metadata
metadata = pd.read_csv("enriched_company_metadata_full.csv") # add the metadata for the companies with over 1000 reviews 

# Ensure 'company_name' exists in both and is string
df['company_name'] = df['company_name'].astype(str)
metadata['company_name'] = metadata['company_name'].astype(str)

# Merge on company_name
df_merged = df.merge(metadata, on='company_name', how='left')

# Save merged file (optional)
#df_merged.to_csv("df_with_company_metadata.csv", index=False)

# Check result
print(df_merged.head())

```
